# 01 — Explore the Synthetic Dataset

This notebook walks through the bundled synthetic menstrual health dataset.

**Everything here is fake.** The data is generated by `synthetic-data/generate.py` so contributors can build and test infrastructure without real participant data. Do not interpret any distribution in this notebook as real prevalence.

Run from the repository root (or adjust `DATA_PATH`).

```python
from pathlib import Path

import pandas as pd

DATA_PATH = Path("..") / "synthetic-data" / "datasets" / "survey_responses_1000.csv"
if not DATA_PATH.exists():
    DATA_PATH = Path("synthetic-data") / "datasets" / "survey_responses_1000.csv"

df = pd.read_csv(DATA_PATH)
df.head()
```

## Dataset shape and field types

Compare what you see against `schema/survey_schema.json` and `schema/data_dictionary.json` — every public dataset in this project should conform to the schema.

```python
print(df.shape)
df.dtypes
```

```python
df[["cycle_length_days", "period_duration_days", "pain_score"]].describe()
```

## Distributions of key numeric fields

```python
axes = df[["cycle_length_days", "period_duration_days", "pain_score"]].hist(figsize=(12, 3), layout=(1, 3), bins=15)
```

## Categorical fields

All categorical fields use controlled vocabularies (see the schema). Controlled categories make validation, translation, and cross-dataset comparison possible.

```python
for column in ["age_band", "setting", "flow_heaviness", "product_access", "healthcare_access"]:
    print(f"--- {column} ---")
    print(df[column].value_counts(normalize=True).round(3))
    print()
```

## Symptoms

`reported_symptoms` is a semicolon-separated list drawn from a controlled symptom vocabulary. Exploding it gives one row per symptom mention.

```python
symptoms = (
    df["reported_symptoms"].fillna("").str.split(";").explode().str.strip()
)
symptoms = symptoms[symptoms != ""]
symptoms.value_counts().plot.barh(title="Symptom mentions (synthetic)")
```
